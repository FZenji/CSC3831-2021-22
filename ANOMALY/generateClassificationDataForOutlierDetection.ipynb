## generate Classification Data For Outlier Detection

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
```

use the code in the next cell to explore how novelty detection (OneClassSVM) and outlier detection methods (LOF, Isolation Forest) perform when you vary:
 - fraction_outliers:  0.001, 0.01, 0.05
 - class_sep:  0,2,4  please see [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html?highlight=make_classification#sklearn.datasets.make_classification) for documentation pn class_sep.
 

```python
fraction_outliers = 0.05
class_sep = 0

n_features = 5

X, CL = make_classification(n_samples=10000, n_classes=2, n_features=n_features, 
    n_redundant=0, random_state=5, weights=[1-fraction_outliers], n_clusters_per_class=2,class_sep=class_sep, flip_y = 0)

## count of each label
unique_elements, counts_elements = np.unique(CL, return_counts=True)
print('count of each class label (0: inliers, 1: outliers)')
print(np.asarray((unique_elements, counts_elements)))
```

 The effect of changing these parameters can be observed in the plot below. Notice that only the first 2 features are plotted (plus the colour-coded class label)

```python
# plot entire X,Y
plt.figure(figsize=(10,6))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=CL)
```
